```python
from natasha import Segmenter, NewsEmbedding, NewsSyntaxParser, Doc
```

```python
emb = NewsEmbedding()
```

```python
segmenter = Segmenter()
```

```python
syntax_parser = NewsSyntaxParser(emb)
```

```python
text = 'Целый день ребята трудились – строили снежную горку во дворе'
```

```python
doc = Doc(text)
```

```python
doc.segment(segmenter)
```

```python
for tocken in doc.tokens:
    print(tocken)
```

```python
doc.parse_syntax(syntax_parser)
```

```python
for tocken in doc.tokens:
    print(tocken)
```
