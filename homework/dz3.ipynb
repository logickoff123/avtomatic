```python
def vector_na_vector(vec1, vec2):
    return sum(i * j for i, j in zip(vec1[0], vec2[0]))

def normal(vec):
    print(vec.matrix)
    result = (vec.matrix[0][0]**2 + vec.matrix[0][1]**2 + vec.matrix[0][2]**2 + vec.matrix[0][3]**2)**0.5 #норма вектора  (sqrt(v1^2 +v2^2 + v3^2))
    return result


class Matrix:
    def __init__(self,matrix):
        self.matrix = matrix
        
    # def eq(self, other_matrix):
    #     if len(self.matrix)!= len(other_matrix) or len(self.matrix[0])!=len(other_matrix[0]):
    #         print ("Размерность попутал")
    #         return 0
    #     for i in range(len(self.matrix)):
    #         for J in range (len(self.matrix[0])):
    #             if self.matrix[i][j]!=other_matrix[i][j]:
    #                 return 0
    #     return 1
    
    def __add__ (self, other):
        resultat = [] # создаем пустой список для результат
        if isinstance(other,int) or isinstance(other,float): # проверка на число 
           for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = [] # создаем пустую строку для новой матрицы 
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице  
                    new_line.append (self.matrix[line][column] + other)  # складываем -_-
                resultat.append(new_line) 
        else:
            for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = [] # создаем пустую строку для новой матрицы
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице  
                    new_line.append (self.matrix[line][column] + other.matrix[line][column] )  
                resultat.append(new_line)
        return Matrix(resultat)
    
    # add, sub, truediv и mul работают по такому же принципу просто меняем знак 
    #сколярка и транспонирование работают по другому сейчас объясню
    
    def __sub__ (self, other):
        resultat = []
        if isinstance(other,int) or isinstance(other,float):
           for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = []
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице  
                    new_line.append (self.matrix[line][column] - other)  
                resultat.append(new_line)
        else:
            for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = []
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице 
                    new_line.append (self.matrix[line][column] - other.matrix[line][column] )  
                resultat.append(new_line)
        return Matrix(resultat)
                
                
    def __truediv__ (self, other):
        resultat = []
        if isinstance(other,int) or isinstance(other,float):
           for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = []
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице  
                    new_line.append (self.matrix[line][column] / other)  
                resultat.append(new_line)
        else:
            for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = []
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице 
                    new_line.append (self.matrix[line][column] / other.matrix[line][column] )  
                resultat.append(new_line)
        return Matrix(resultat)            
    
    # сколярка делаеться следующим образом :строка на столбец
    # поэтому сначала проходимся по строчкам в первой потом по столбцам во второй
    # однако строчку на строку мы то умножаем, но не забываем что и строка и столбец
    # состоят из элементов, поэтому нам понадобиться тут три цикла 
    
    
    def __matmul__(self,other):
        resultat = [] # тут по классике создаем список для результата
        for line in range(len(self.matrix)):# прогоняем по строчкам матрицу первую
            new_line = [] # новую матрицу сюда запихнем
            for column in range(len(other.matrix[0])):#прогоняем по столбцам матрицу вторую
                skolar_peremen = 0 # переменная для сколярного произведения  
                for i in range(len(self.matrix[0])): # прогоняем по элементам строки перой матрицы и столбца второй матрицы 
                    skolar_peremen += self.matrix[line][i] * other.matrix[i][column] # сколярка 
                new_line.append(skolar_peremen)
        resultat.append(new_line)
    
        return Matrix(resultat)
    
    def __mul__(self,other):
        resultat = []
        if isinstance(other,int) or isinstance(other,float):
           for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = []
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице  
                    new_line.append (self.matrix[line][column] * other)  
                resultat.append(new_line)
        else:
            for line in range(len(self.matrix)):# прогняемся по строчкам
                new_line = []
                for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице  
                    new_line.append (self.matrix[line][column] * other.matrix[line][column] )  
                resultat.append(new_line)
        return Matrix(resultat)  
    
    # меняем строки и столбцы местами, так как транспонированная матрица начинаеться со столбца, которая была строчкой у первой , то сначала прогоняемся по столбцам
    def T(self):
        resultat = [] # и снова создаем список для результата
        for column in range(len(self.matrix[0])): # прогоняемя по столбцам в строчке начиная -> matrix[0]- первый элемнт в матрице
            new_line = []
            for line in range(len(self.matrix)): # прогоняем по строкам 
                new_line.append(self.matrix[line][column]) # добовляем элемент из столбца в новую строчку 
            resultat.append (new_line)
        return Matrix(resultat)
    
    def empty_vectors(self): # создание пустых списков для каждого вектора 
        v1, v2, v3, v4 = [],[],[],[]
        for i in self.matrix: # добавим элементы из каждого столбца в вектор который
            v1.append(i[0]) 
            v2.append(i[1])
            v3.append(i[2])
            v4.append(i[3])
        return [v1,], [v2,], [v3,], [v4,]


    def basises(self):
        v1 = self.empty_vectors()[0]
        v2 = self.empty_vectors()[1]
        v3 = self.empty_vectors()[2]
        v4 = self.empty_vectors()[3]

        f1 = v1

        f2 = (Matrix(v2) - (Matrix(f1)*(vector_na_vector(v2,f1) / vector_na_vector(f1,f1)))).matrix

        f3 = (Matrix(v3) - (Matrix(f1)*(vector_na_vector(v3,f1) / vector_na_vector(f1,f1))) -(Matrix(f2)*(vector_na_vector(v3,f2) / vector_na_vector(f2,f2)))).matrix
        
        f4 = (Matrix(v4) - (Matrix(f1)*(vector_na_vector(v4,f1) / vector_na_vector(f1,f1))) -(Matrix(f2)*(vector_na_vector(v4,f2) / vector_na_vector(f2,f2))) -(Matrix(f3)*(vector_na_vector(v4,f3) / vector_na_vector(f3,f3)))).matrix
        return Matrix(f1), Matrix(f2), Matrix(f3), Matrix(f4)
    
    def orthogonal(self):
        res = []
        for f in self.basises():
            res.append((f/normal(f)).matrix)
        return(res)
```
